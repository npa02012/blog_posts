```python
import time
import matplotlib.pyplot
import random
import numpy as np


def calculate_complexity(fun, max_time=5):
    '''
    Parameters:
    fun : Funciton, which has an argument n, that will be tested
    max_time : Maximum time to test for (seconds)
    
    Returns:
    [n_it, times] : n_it and times are lists of equal length. n_it represents
                        the n value passed into fun, and times represents how
                        long fun took to run for that particular value of n
    '''
    
    n_it = []
    times = []
    
    def run_fun(fun, n, n_it, times):
        # Record size of request
        n_it.append(n)
        
        # Run and time request
        s = time.time()
        fun(n)
        time_taken = time.time() - s
        
        # Record and return time taken
        times.append(time_taken)
        return(time_taken)
    
    n = 1
    total_s = time.time()
    while True:
        
        # Run the function for for size n
        time_taken = run_fun(fun, n, n_it, times)
        
        # Update size
        if time_taken < .5:
            n *= 2
        else:
            n = int(n * 1.3)
        
        # Break if taking too long
        if (time.time() - total_s) > max_time:
            print('Ending at len(n_it) = ' + str(len(n_it))
                 + ' and max(n_it) = ' + str(max(n_it)))
            break
            
        # Break the loop if n is huge
        max_n = 100 * 1000000
        if n > (max_n):
            print('Ending because n > ' + str(max_n))
            break
    
    return [n_it, times]


# - - - - - - - - - - - - - - - - - - - - - - - - 
# Accessing Array : O(1)

def list_access(n):
    L = range(n)
    tmp = L[random.choice(range(n))]

# - - - - - - - - - - - - - - - - - - - - - - - - 
# Sorting a list : supposed to be O(n log(n)))

def list_sort(n):
    L = random.sample(range(n), n)
    L.sort()
    
# - - - - - - - - - - - - - - - - - - - - - - - - 
# Getting min of a list : O(n)

def list_min(n):
    L = random.sample(range(n), n)
    tmp = min(L)
    
# - - - - - - - - - - - - - - - - - - - - - - - - 
# Getting max of a list : O(n)

def list_max(n):
    L = random.sample(range(n), n)
    tmp = max(L)

# - - - - - - - - - - - - - - - - - - - - - - - - 
# Iterating over a list : O(n)

def list_iterate(n):
    L = range(n)
    for l in L:
        pass

# - - - - - - - - - - - - - - - - - - - - - - - - 
# Iterate over 2 lists of size n : O(n^2)
def list_two_iterations(n):
    L = range(n)
    L2 = range(n)
    for l in L:
        for l2 in L2:
            pass
        
# - - - - - - - - - - - - - - - - - - - - - - - - 
# Iterate over 3 lists of size n : O(n^3)
def list_three_iterations(n):
    L = range(n)
    for l in L:
        for l2 in L:
            for l3 in L:
                pass
            
# - - - - - - - - - - - - - - - - - - - - - - - - 
# Define all functions
all_fun = {}
all_fun['list_access'] = calculate_complexity(list_access)
all_fun['list_iterate'] = calculate_complexity(list_iterate)
all_fun['list_sort'] = calculate_complexity(list_sort)
all_fun['list_min'] = calculate_complexity(list_min)
all_fun['list_max'] = calculate_complexity(list_min)
all_fun['list_two_iterations'] = calculate_complexity(list_two_iterations)
all_fun['list_three_iterations'] = calculate_complexity(list_three_iterations)
```

```python
def plot_complexities(plot_keys):
    for k in plot_keys:
        matplotlib.pyplot.plot(all_fun[k][0], all_fun[k][1])


    # Add legend and show
    matplotlib.pyplot.legend(plot_keys)
    matplotlib.pyplot.show()
```
